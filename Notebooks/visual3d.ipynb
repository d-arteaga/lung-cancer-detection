```python
import SimpleITK as sitk
import matplotlib.pyplot as plt

%matplotlib inline

data_path = 'C:/Users/darte/Documents/Projects/lung-cancer-detection/Data/LIDC-IDRI/LIDC-IDRI-0001/01-01-2000-NA-NA-30178/3000566.000000-NA-03192'
reader = sitk.ImageSeriesReader()
dicom_names = reader.GetGDCMSeriesFileNames(data_path)  
reader.SetFileNames(dicom_names)
image = reader.Execute()
```

```python
img_array = sitk.GetArrayFromImage(image)

axial = img_array[img_array.shape[0] // 2, :, :]

coronal = img_array[:, img_array.shape[1] // 2, :]

sagittal = img_array[:, :, img_array.shape[2] // 2]

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

fig.suptitle("CT Scan Slices in Different Planes: Patient 1", fontsize=18)

axes[0].imshow(axial, cmap='gray')
axes[0].set_title("Axial Slice")
axes[1].imshow(coronal, cmap='gray')
axes[1].set_title("Coronal Slice")
axes[2].imshow(sagittal, cmap='gray')
axes[2].set_title("Sagittal Slice")
plt.show()
```

```python
# Update the data path for patient 2
data_path_patient_2 = 'C:/Users/darte/Documents/Projects/lung-cancer-detection/Data/LIDC-IDRI/LIDC-IDRI-0002/01-01-2000-NA-NA-98329/3000522.000000-NA-04919'

# Read the DICOM series for patient 2
reader.SetFileNames(reader.GetGDCMSeriesFileNames(data_path_patient_2))
image_patient_2 = reader.Execute()

# Convert the image to a numpy array
img_array_patient_2 = sitk.GetArrayFromImage(image_patient_2)

# Reslice the image in different planes
# Axial (default)
axial_patient_2 = img_array_patient_2[img_array_patient_2.shape[0] // 2, :, :]

# Coronal (slice along the y-axis)
coronal_patient_2 = img_array_patient_2[:, img_array_patient_2.shape[1] // 2, :]

# Sagittal (slice along the x-axis)
sagittal_patient_2 = img_array_patient_2[:, :, img_array_patient_2.shape[2] // 2]

# Plot the slices
fig_patient_2, axes_patient_2 = plt.subplots(1, 3, figsize=(15, 5))

fig_patient_2.suptitle("CT Scan Slices in Different Planes: Patient 2", fontsize=18)

axes_patient_2[0].imshow(axial_patient_2, cmap='gray')
axes_patient_2[0].set_title("Axial Slice")
axes_patient_2[1].imshow(coronal_patient_2, cmap='gray')
axes_patient_2[1].set_title("Coronal Slice")
axes_patient_2[2].imshow(sagittal_patient_2, cmap='gray')
axes_patient_2[2].set_title("Sagittal Slice")
plt.show()
```

```python
import pyvista as pv
import numpy as np

new_img_array = np.transpose(img_array, (2, 1, 0))

volume = pv.wrap(new_img_array)

plotter = pv.Plotter()
plotter.add_title("CT Scan 3D Render: Patient 1")
plotter.add_volume(volume, cmap="gray", opacity="sigmoid")  
plotter.show()
```
